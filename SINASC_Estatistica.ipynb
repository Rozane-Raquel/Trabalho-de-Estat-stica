```python
import pandas as pd

# lendo a base de dados
df=pd.read_csv('/content/drive/MyDrive/Estatistica computacional/SINASC_2024.csv', sep=';')

#evita que alguma coluna não apareça
pd.set_option('display.max_columns', None)





```

```python
#  remover colunas irrelevantes
df = df.drop(columns=['ORIGEM', 'CODESTAB', 'DTCADASTRO','NUMEROLOTE','VERSAOSIST',
                      'DTRECEBIM', 'DIFDATA','DTRECORIGA', 'TPFUNCRESP','TPDOCRESP',
                      'OPORT_DN','STDNEPIDEM','STDNNOVA', 'DTDECLARAC'])
```

```python
# exibir 20 linhas do dataframe
df.sample(20)
```

```python


# mostrar resumo das colunas
df.info()

```

```python
# estatísticas descritivas
df['IDADEMAE'].describe()
```

```python
# transformando o tipo dos dados de IDADEMAE para int
df['IDADEMAE'] = df['IDADEMAE'].astype('Int64')
```

```python
#coluna categórica nova
mapa_locnasc = {
    1: 'Hospital',
    2: 'Outros estabelecimentos de saúde',
    3: 'Domicílio',
    4: 'Outros',
    5: 'Aldeia Indígena'
}


df['LOCNASC_DESC'] = df['LOCNASC'].map(mapa_locnasc)
df[['LOCNASC', 'LOCNASC_DESC']].head(3)
```

```python
df['LOCNASC_DESC'].value_counts().plot(kind='bar')
```

```python
#coluna categórica nova
mapa_estciv = {
    1: 'Solteira',
    2: 'Casada',
    3: 'Viúva',
    4: 'Separada/divorciada',
    5: 'União estável',
    9: 'Ignorada'
}

df['ESTCIVMAE_DESC'] = df['ESTCIVMAE'].map(mapa_estciv)

# exibindo a nova coluna
df[['ESTCIVMAE', 'ESTCIVMAE_DESC']].head(20)
```

```python
df['ESTCIVMAE_DESC'].value_counts().plot(kind='bar')
```
